```python
pressure_data = pd.read_csv('D:/IMS Proschool/Business Analytics/Time series/Case Studies - Questions with Solutions-20190724/Case 2/pressure.csv')
```

```python
#convert to month and year
dates = pd.date_range(start='2018-01-01', freq='D', periods=len(pressure_data))
import calendar
pressure_data['Month'] = dates.month
pressure_data['Month'] = pressure_data['Month'].apply(lambda x : calendar.month_abbr[x])
pressure_data['Year'] = dates.year
```

```python
#create month and year
pressure_data = pressure_data[['Month','Year','MAIN_FILTER_IN_PRESSURE']]
pressure_data.head()
```

```python
#rename columns for better use
pressure_data.columns = ['Month','Year','Pressure']
pressure_data.head()
```

```python
#set index as dates
pressure_data.set_index(dates, inplace=True)
pressure_data.head()
```

```python
#extract the time series
pressure = pressure_data[['Pressure']]
pressure.head()
```

```python
#plot the time series
plt.figure(figsize=(15,5))
plt.plot(pressure)
plt.xlabel('Months')
plt.ylabel('Pressure')
plt.show()
```

```python
#rolling statistics
rollmean = pressure.rolling(window=12).mean()
rollstd = pressure.rolling(window=12).std()

#plot rolling statistics
plt.figure(figsize=(15,5))
plt.plot(pressure, label='Original')
plt.plot(rollmean, label='Rolling Mean')
plt.plot(rollstd, label='Rolling Std')
plt.legend(loc='best')
plt.show()
```

```python
telcom = pd.read_csv('D:/IMS Proschool/Business Analytics/Classification/Case Studies - Questions with Solutions-20190724/Decision Trees/Case 2/churn.csv')
```

```python
#replace spaces with null values
telcom['TotalCharges'] = telcom['TotalCharges'].replace(" ", np.nan)
```

```python
#replace no internet service to no for the following columns:
replace_cols = ['OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']
for i in replace_cols:
    telcom[i] = telcom[i].replace({'No internet service':'No'})
```

```python
#change numeric to categorical column
telcom['SeniorCitizen'] = telcom['SeniorCitizen'].replace({1:'Yes',0:'No'})
```

```python
#change tenure to categorical columns / binning and assign it to a new column tenure_group
def tenure_lab(telcom):
    if telcom["tenure"] <= 12:
        return "Tenure_0-12"
    elif (telcom["tenure"] > 12) & (telcom["tenure"] <= 24 ):
        return "Tenure_12-24"
    elif (telcom["tenure"] > 24) & (telcom["tenure"] <= 48) :
        return "Tenure_24-48"
    elif (telcom["tenure"] > 48) & (telcom["tenure"] <= 60) :
        return "Tenure_48-60"
    elif telcom["tenure"] > 60 :
        return "Tenure_gt_60"
telcom["tenure_group"] = telcom.apply(lambda telcom:tenure_lab(telcom), axis=1)
```

```python
#Binary columns with 2 values
bin_cols   = telcom.nunique()[telcom.nunique() == 2].keys().tolist()

#Label encoding Binary columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in bin_cols :
    telcom[i] = le.fit_transform(telcom[i])
```
